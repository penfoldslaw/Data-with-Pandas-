## Are Franchises more successful?

__Franchise (True) vs. Stand-alone (False): Movies made__

```python
df_best2 = df.copy()
```

```python
collection = df_best2.belongs_to_collection.notna()
```

```python
collection.value_counts()
```

__Franchise (True) vs. Stand-alone (False):Average Revenue__

```python
df_best2.groupby(collection).revenue_musd.mean()
```

__Franchise (True) vs. Stand-alone (False): Return on Investment / Profitability__

```python
df_best2["ROI"] = df_best2.revenue_musd/(df_best2.budget_musd)
```

```python
df_best2.groupby(collection).ROI.median()
```

 __Franchise (True) vs. Stand-alone (False): Average Budget__

```python
df_best2.groupby(collection).budget_musd.mean()
```

__Franchise (True) vs. Stand-alone (False): All the difference__

```python
df_best2.groupby(collection).agg({"budget_musd": "mean", "revenue_musd": "mean", "vote_average": "mean",
                            "popularity": "mean", "ROI":"median", "vote_count":"mean"})
```

## Most Successful Franchises

```python
df_best2.belongs_to_collection.dropna().value_counts().head(10)
```

```python
prop1 = df_best2.belongs_to_collection
```

```python
collection1 = df_best2.groupby(prop1).agg({'title':'count',"budget_musd": ["sum", "mean"], "revenue_musd": ["sum", "mean"], "vote_average": "mean", 
                                   "popularity": "mean","ROI":"median", "vote_count":"mean"})
```

```python
collection1.nlargest(20, ("title", "count"))
```

```python
collection1.nlargest(20, ("revenue_musd", "mean"))
```

```python
collection1.nlargest(20, ("budget_musd", "mean"))
```

```python
collection1[collection1[("vote_count", "mean")] >=1000].nlargest(20, ("vote_average", "mean"))
```

## Most Successful Directors

```python
df_best2.director.dropna().value_counts().head(10)
```

```python
direct1 = df_best2.groupby('director').agg(Total_count = ('title','count'), Total_Revenue = ('revenue_musd','sum'))
```

```python
plt.figure(figsize = (12, 8))
direct1.Total_Revenue.nlargest(10).plot(kind='bar', fontsize = 15)
plt.title("Total Revenue",fontsize = 20)
plt.ylabel("Revenue (in MUSD)", fontsize = 15)
plt.show()
```
